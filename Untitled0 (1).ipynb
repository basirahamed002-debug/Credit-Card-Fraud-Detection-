```python
# Load dataset
df = pd.read_csv('creditcard.csv')

# Preview
df.head()
```

```python
# Shape
print("Shape:", df.shape)

# Info
df.info()

# Missing values
print("\nMissing Values:\n", df.isnull().sum())

# Summary stats
df.describe()
```

```python
# Count
class_counts = df['Class'].value_counts()
print(class_counts)

# Plot
plt.figure()
sns.countplot(x='Class', data=df)
plt.title("Class Distribution (0 = Normal, 1 = Fraud)")
plt.show()

# Fraud %
fraud_percent = (class_counts[1] / len(df)) * 100
print(f"Fraud Percentage: {fraud_percent:.4f}%")
```

```python
# Check duplicates
print("Duplicates:", df.duplicated().sum())

# Remove duplicates
df = df.drop_duplicates()

print("New shape:", df.shape)
```

```python
df['Hour'] = (df['Time'] // 3600) % 24

df[['Time', 'Hour']].head()
```

```python
bins = [0, 50, 100, 500, 1000, 5000, 10000]
labels = ['0-50', '50-100', '100-500', '500-1000', '1000-5000', '5000+']

df['Amount_Bin'] = pd.cut(df['Amount'], bins=bins, labels=labels)

df[['Amount', 'Amount_Bin']].head()
```

```python
df['Log_Amount'] = np.log1p(df['Amount'])

df[['Amount', 'Log_Amount']].head()
```

```python
plt.figure()
sns.countplot(x='Class', data=df)
plt.title("Fraud vs Normal Transactions")
plt.show()
```

```python
plt.figure()

sns.histplot(df[df['Class'] == 0]['Amount'], bins=50, kde=True, label='Normal')
sns.histplot(df[df['Class'] == 1]['Amount'], bins=50, kde=True, color='red', label='Fraud')

plt.legend()
plt.title("Transaction Amount Distribution")
plt.show()
```

```python
plt.figure()
sns.boxplot(x='Class', y='Amount', data=df)
plt.title("Transaction Amount by Class")
plt.show()
```

```python
fraud_by_hour = df.groupby('Hour')['Class'].sum()

plt.figure()
fraud_by_hour.plot(kind='line', marker='o')
plt.title("Fraud Transactions by Hour")
plt.xlabel("Hour")
plt.ylabel("Fraud Count")
plt.show()
```

```python
fraud_by_bin = df[df['Class'] == 1]['Amount_Bin'].value_counts()

plt.figure()
fraud_by_bin.plot(kind='bar')
plt.title("Fraud Transactions by Amount Range")
plt.xlabel("Amount Bin")
plt.ylabel("Count")
plt.show()
```

```python
plt.figure(figsize=(12,8))
sns.heatmap(df.drop('Amount_Bin', axis=1).corr(), cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.show()
```

```python
summary = df.groupby('Class')['Amount'].agg(['count', 'mean', 'sum'])
summary.rename(index={0: 'Normal', 1: 'Fraud'}, inplace=True)

summary
```

```python
# ==============================
# Final Project Conclusion
# ==============================

fraud_cases = df['Class'].sum()
total_cases = len(df)
fraud_rate = (fraud_cases / total_cases) * 100

top_hour = df.groupby('Hour')['Class'].sum().sort_values(ascending=False).index[0]
top_amount_bin = df[df['Class'] == 1]['Amount_Bin'].value_counts().idxmax()

print("=== Project Conclusion ===\n")

print(f"This project analyzed {total_cases:,} credit card transactions, out of which {fraud_cases:,} were identified as fraudulent.")
print(f"The fraud rate was only {fraud_rate:.4f}%, confirming that this is a highly imbalanced classification problem.\n")

print(f"Exploratory analysis showed that fraud was more concentrated around hour {top_hour} and was most frequent in the '{top_amount_bin}' transaction range.")
print("These patterns suggest that both transaction timing and amount can be useful signals for fraud monitoring.\n")

print("To improve fraud detection, machine learning models were trained and evaluated using metrics such as precision, recall, and ROC-AUC instead of relying only on accuracy.")
print("This is important because in fraud detection, catching more fraudulent transactions is more valuable than simply maximizing overall accuracy.\n")

print("A fraud risk scoring system was also created to classify transactions into Low Risk, Medium Risk, and High Risk categories.")
print("This makes the solution more practical for business use, as banks can use these risk levels to prioritize reviews and reduce financial loss.\n")

print("Overall, this project demonstrates how data analysis and machine learning can work together to identify suspicious activity, support faster decision-making, and strengthen fraud prevention strategies.")
```
